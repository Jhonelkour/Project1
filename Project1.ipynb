# Data Visualization Section

## 1. Weather Conditions Distribution

```python
# Weather Conditions Bar Chart
plt.figure(figsize=(12, 6))
weather_counts = df['Weather_Condition'].value_counts()

plt.subplot(1, 2, 1)
weather_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Weather Conditions Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Weather Condition', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)

# Pie chart for weather conditions
plt.subplot(1, 2, 2)
plt.pie(weather_counts.values, labels=weather_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Weather Conditions Distribution (%)', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

## 2. Temperature Analysis

```python
# Temperature Distribution and Box Plot
plt.figure(figsize=(15, 5))

# Temperature histogram
plt.subplot(1, 3, 1)
plt.hist(df['Temp_C'], bins=30, color='orange', alpha=0.7, edgecolor='black')
plt.title('Temperature Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Temperature (°C)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.3)

# Temperature vs Dew Point
plt.subplot(1, 3, 2)
plt.scatter(df['Temp_C'], df['Dew Point Temp_C'], alpha=0.6, color='green')
plt.title('Temperature vs Dew Point', fontsize=14, fontweight='bold')
plt.xlabel('Temperature (°C)', fontsize=12)
plt.ylabel('Dew Point Temperature (°C)', fontsize=12)
plt.grid(True, alpha=0.3)

# Temperature by Weather Condition (Box Plot)
plt.subplot(1, 3, 3)
sns.boxplot(data=df, x='Weather_Condition', y='Temp_C')
plt.title('Temperature by Weather Condition', fontsize=14, fontweight='bold')
plt.xlabel('Weather Condition', fontsize=12)
plt.ylabel('Temperature (°C)', fontsize=12)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## 3. Wind Speed and Visibility Analysis

```python
# Wind Speed and Visibility Analysis
plt.figure(figsize=(15, 5))

# Wind Speed Distribution
plt.subplot(1, 3, 1)
plt.hist(df['Wind Speed_km/h'], bins=25, color='lightblue', alpha=0.7, edgecolor='black')
plt.title('Wind Speed Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Wind Speed (km/h)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.3)

# Visibility Distribution
plt.subplot(1, 3, 2)
plt.hist(df['Visibility_km'], bins=25, color='lightgreen', alpha=0.7, edgecolor='black')
plt.title('Visibility Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Visibility (km)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.3)

# Wind Speed vs Visibility Scatter Plot
plt.subplot(1, 3, 3)
plt.scatter(df['Wind Speed_km/h'], df['Visibility_km'], alpha=0.6, color='purple')
plt.title('Wind Speed vs Visibility', fontsize=14, fontweight='bold')
plt.xlabel('Wind Speed (km/h)', fontsize=12)
plt.ylabel('Visibility (km)', fontsize=12)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 4. Humidity and Pressure Analysis

```python
# Humidity and Pressure Analysis
plt.figure(figsize=(12, 8))

# Relative Humidity by Weather Condition
plt.subplot(2, 2, 1)
sns.violinplot(data=df, x='Weather_Condition', y='Rel Hum_%')
plt.title('Relative Humidity by Weather Condition', fontsize=14, fontweight='bold')
plt.xlabel('Weather Condition', fontsize=12)
plt.ylabel('Relative Humidity (%)', fontsize=12)
plt.xticks(rotation=45)

# Pressure Distribution
plt.subplot(2, 2, 2)
plt.hist(df['Press_kPa'], bins=30, color='red', alpha=0.7, edgecolor='black')
plt.title('Atmospheric Pressure Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Pressure (kPa)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.3)

# Humidity vs Pressure
plt.subplot(2, 2, 3)
plt.scatter(df['Rel Hum_%'], df['Press_kPa'], alpha=0.6, color='brown')
plt.title('Humidity vs Pressure', fontsize=14, fontweight='bold')
plt.xlabel('Relative Humidity (%)', fontsize=12)
plt.ylabel('Pressure (kPa)', fontsize=12)
plt.grid(True, alpha=0.3)

# Pressure by Weather Condition
plt.subplot(2, 2, 4)
sns.boxplot(data=df, x='Weather_Condition', y='Press_kPa')
plt.title('Pressure by Weather Condition', fontsize=14, fontweight='bold')
plt.xlabel('Weather Condition', fontsize=12)
plt.ylabel('Pressure (kPa)', fontsize=12)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## 5. Correlation Analysis

```python
# Correlation Heatmap
plt.figure(figsize=(12, 8))

# Select only numerical columns for correlation
numerical_cols = df.select_dtypes(include=[np.number]).columns
correlation_matrix = df[numerical_cols].corr()

# Create heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=0.5, cbar_kws={"shrink": .8})
plt.title('Correlation Matrix of Weather Variables', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

# Print strong correlations
print("Strong Correlations (> 0.7 or < -0.7):")
print("="*50)
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr_value = correlation_matrix.iloc[i, j]
        if abs(corr_value) > 0.7:
            col1 = correlation_matrix.columns[i]
            col2 = correlation_matrix.columns[j]
            print(f"{col1} vs {col2}: {corr_value:.3f}")
```

## 6. Advanced Statistical Analysis

```python
# Pair Plot for Key Variables
key_variables = ['Temp_C', 'Rel Hum_%', 'Wind Speed_km/h', 'Visibility_km', 'Press_kPa']
plt.figure(figsize=(15, 12))

# Create pair plot
sns.pairplot(df[key_variables + ['Weather_Condition']], 
             hue='Weather_Condition', 
             diag_kind='hist',
             plot_kws={'alpha': 0.6})
plt.suptitle('Pair Plot of Weather Variables by Condition', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```
